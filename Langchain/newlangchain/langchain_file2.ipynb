### Learning Models Integration With OpenAI, Google Gemini and GROQ

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="llama-3.3-70b-versatile")
response = model.invoke("Why do parrots talk?")

for chunk in model.stream("Why do parrots talk?"):
    print(chunk.content, end="", flush=True)
    #Forces immediate printing (real-time feel)
    
```

```python
for chunk in model.stream("Write a short stoy about AI"):
    print(chunk.content, end="")
```

```python
questions = [
    "What is AI?",
    "What is ML?",
    "What is Deep Learning?"
]

responses = model.batch(questions)

for res in responses:
    print(res.content)
    
```
