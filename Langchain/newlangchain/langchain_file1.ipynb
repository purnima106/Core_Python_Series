```python
import os
from dotenv import load_dotenv
load_dotenv(dotenv_path="C:\\Users\\Purnima\\OneDrive\\Desktop\\Learning_Everyday\\Langchain\\.env") 
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent

# 1. Connect to Groq instead of OpenAI
llm = ChatGroq(
    model="llama-3.3-70b-specdec", 
    groq_api_key=os.getenv("GROQ_API_KEY")
)

def get_weather(city:str)-> str:
    """Get weather for a city"""
    return f"The weather of the {city} is sunny"

# 2. Pass the 'llm' object to create_agent
agent = create_agent(
    model=llm, 
    tools=[get_weather],
    system_prompt="You are a helpful assistant."
)

# 3. Invoke it exactly the same way
response = agent.invoke({
    "messages": [
        {"role":"user", "content":"What is the weather in Mumbai"}
    ]
})

print(response["messages"][-1].content)
```
